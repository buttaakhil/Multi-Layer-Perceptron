```python
import pandas as pd
```

```python
dataset=pd.read_csv("Churn_Modelling.csv")
```

```python
dataset.head(3)
```

```python
dataset.drop(columns=["RowNumber","CustomerId","Surname","Geography","Gender"],inplace=True)
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
x=pd.DataFrame(ss.fit_transform(x),columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
import tensorflow
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann=Sequential()
```

```python
ann.add(Dense(units=6,input_dim=8,activation="relu"))
ann.add(Dense(units=4,activation="relu"))
ann.add(Dense(units=2,activation="relu"))
ann.add(Dense(units=1,activation="sigmoid"))
```

```python
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
ann.fit(x_train,y_train,batch_size=100,epochs=50)
```

```python
prd1=ann.predict(x_train)
prd1_data=[]
for i in prd1 :
  if i>0.5 :
    prd1_data.append(1)
  else :
    prd1_data.append(0)
```

```python
prd2=ann.predict(x_test)
prd2_data=[]
for i in prd2 :
  if i>0.5 :
    prd2_data.append(1)
  else :
    prd2_data.append(0)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_train,prd1_data)*100, accuracy_score(y_test,prd2_data)*100
```
